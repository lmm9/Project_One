```python
import pandas as pd
import json
import requests
```

```python
# set up lists to hold reponse info
report_date = []
status=[]
gender = []
age_group=[]
race_ethnicity=[]
hospital=[]
icu=[]
death=[]
med_cond=[]
posted=[]
onset=[]

offset=0
for j in range(75):
    offset= j*50000  
    
    # Build partial query URL
    url = f'https://data.cdc.gov/resource/vbim-akqf.json?$limit=50000&$offset={offset}'
    response = requests.get(url).json()

    # Loop through the list of data and perform a request for data on each
    for i in range(0, 50000):
   
        try:
            report_date.append(response[i]['cdc_report_dt'])
            status.append(response[i]['current_status'])
            gender.append(response[i]['sex'])
            age_group.append(response[i]['age_group'])
            race_ethnicity.append(response[i]['race_ethnicity_combined'])
            hospital.append(response[i]['hosp_yn'])
            icu.append(response[i]['icu_yn'])
            death.append(response[i]['death_yn'])
            med_cond.append(response[i]['medcond_yn'])
            posted.append(response[i]['pos_spec_dt'])
            onset.append(response[i]['onset_dt'])

        except:
            pass   
```

```python
# create a data frame from covid data
covid_dict = {
    "report_date":report_date,
    "status":status,
    "gender":gender,
    "age_group":age_group,
    "race_ethnicity":race_ethnicity,
    "hospital":hospital,
    "icu":icu,
    "death":death,
    "med_cond":med_cond,
}
covid_data = pd.DataFrame(covid_dict)
covid_data
```

```python
# Export file as a CSV, without the Pandas index, but with the header
covid_data.to_csv("Output/covid_dataframe.csv", index=False, header=True)
```
