```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read CSV
# Store filepath in a variable
csv_path = "Output/covid_week.csv"
# Read our Data file with the pandas library
covid_week_data = pd.read_csv(csv_path, encoding="utf-8")
covid_week_data.set_index("week", inplace=True)
covid_week_data
```

```python
covid_gender_df=covid_week_data[["Male", "Female", "Unknown Gender"]]
covid_gender_df.head()
```

```python
covid_agegroup_df=covid_week_data[["0 - 9 Years","10 - 19 Years","20 - 29 Years","30 - 39 Years","40 - 49 Years",
                                 "50 - 59 Years","60 - 69 Years","70 - 79 Years","80+ Years","Unknown Age Group"]]
covid_agegroup_df.head()
```

```python
covid_race_df=covid_week_data[["American Indian/Alaska Native","Asian","Black", "Hispanic/Latino",
                             "Multiple/Other", "Native Hawaiian/Other Pacific Islander", "White", "Unknown Race"]]
covid_race_df.head()
```

```python
covid_death_df=covid_week_data[["Death-Yes", "Death-No", "Death-Unknown"]]
covid_death_df.head()
```

```python
# Use DataFrame.plot() in order to create a chart of the data
covid_gender_sum=covid_gender_df.sum()
colors = ["blue","red","lightgray"]
covid_gender_sum.plot(kind="pie",colors=colors,autopct="%1.1f%%")

# Set a title for the chart
plt.title("Total U.S. COVID-19 Cases by Gender")
plt.axis("equal")
plt.tight_layout()
plt.show()
plt.savefig("Output/covid_gender_total.png")
```

```python
# Use DataFrame.plot() in order to create a chart of the data
covid_gender_df.plot(kind="line",figsize=(10,5))

# Set a title for the chart
plt.title("U.S. COVID-19 Cases by Gender")
plt.xlabel("Week")
plt.ylabel("# Cases")
plt.tight_layout()
plt.show()
plt.savefig("Output/covid_gender_time.png")
```

```python
# Use DataFrame.plot() in order to create a chart of the data
covid_agegroup_sum=covid_agegroup_df.sum()
covid_agegroup_sum.plot(kind="pie", autopct="%1.1f%%")

# Set a title for the chart
plt.title("Total U.S. COVID-19 Cases by Age Group")
plt.axis("equal")
plt.tight_layout()
plt.show()
plt.savefig("Output/covid_agegroup_total.png")
```

```python
# Use DataFrame.plot() in order to create a chart of the data
covid_agegroup_df.plot(kind="line", figsize=(10,5))

# Set a title for the chart
plt.title("U.S. COVID-19 Cases by Age Group")
plt.tight_layout()
plt.show()
plt.savefig("Output/covid_agegroup_time.png")
```

```python
# Use DataFrame.plot() in order to create a chart of the data
covid_race_sum=covid_race_df.sum()
covid_race_sum.plot(kind="pie", autopct="%1.1f%%")

# Set a title for the chart
plt.title("Total U.S. COVID-19 Cases by Age Group")
plt.axis("equal")
plt.tight_layout()
plt.show()
plt.savefig("Output/covid_race_total.png")
```

```python
# Use DataFrame.plot() in order to create a chart of the data
covid_race_df.plot(kind="line", figsize=(10,5))

# Set a title for the chart
plt.title("U.S. COVID-19 Cases by Race")
plt.tight_layout()
plt.show()
plt.savefig("Output/covid_race_time.png")
```

```python
# Use DataFrame.plot() in order to create a chart of the data
covid_death_sum=covid_death_df.sum()
covid_death_sum.plot(kind="pie", autopct="%1.1f%%")

# Set a title for the chart
plt.title("Total U.S. COVID-19 Cases by Age Group")
plt.axis("equal")
plt.tight_layout()
plt.show()
plt.savefig("Output/covid_death_total.png")
```

```python
# Use DataFrame.plot() in order to create a chart of the data
covid_death_df.plot(kind="line", figsize=(10,5))

# Set a title for the chart
plt.title("U.S. COVID-19 Deaths")
plt.tight_layout()
plt.show()
plt.savefig("Output/covid_death_time.png")
```
